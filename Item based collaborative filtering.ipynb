```python
import pandas as pd
r_cols = ['user_id', 'movie_id', 'rating']
ratings=pd.read_csv("C:/Users/manavgakhar/Downloads/Sundog ML/DataScience-Python3/ml-100k/u.data", sep='\t', names=r_cols,usecols=range(3),encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('C:/Users/manavgakhar/Downloads/Sundog ML/DataScience-Python3/ml-100k/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)

```

```python
ratings.head()
```

```python
userRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
userRatings.head()
```

```python
corrMatrix= userRatings.corr()
```

```python
corrMatrix.head()
```

```python
corrMatrix=userRatings.corr(method='pearson',min_periods=200)
```

```python
corrMatrix
```

```python
myratings=userRatings.loc[0].dropna()
```

```python
len(myratings)
```

```python
simCandidates=pd.Series()
for i in range(0,len(myratings)):
    sims= corrMatrix[myratings.index[i]].dropna()
    if(myratings[i]<=2.5):
        sims=sims.map(lambda x:x*0.25)
    else:
        sims=sims.map(lambda x:x*myratings[i])
    simCandidates=simCandidates.append(sims)
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
```

```python
simCandidates=simCandidates.drop(myratings.index)
```
